```python
config = {
    # "env_name": "SpaceInvadersNoFrameskip-v4",
    "env_name": "ALE/Breakout-v5",
    "num_envs": 32,
    "total_timesteps": int(10e6),
    "seed": 0,    
}

# There already exists an environment generator
# that will make and wrap atari environments correctly.
# Here we are also multi-worker training (n_envs=8 => 8 environments)
env = make_atari_env(config["env_name"], n_envs=config["num_envs"], seed=config["seed"]) #PongNoFrameskip-v4

print("ENV ACTION SPACE: ", env.action_space.n)

# Frame-stacking with 4 frames
env = VecFrameStack(env, n_stack=4)

log_dir = f'logs/{config["env_name"]}'
os.makedirs(log_dir, exist_ok=True)

# Set up the monitor to log training information
# env = Monitor(env, log_dir)

# Set up the callback to save the model
checkpoint_callback = CheckpointCallback(save_freq=100000, save_path=log_dir, name_prefix=f'{config["env_name"]}_ckpt')

# Create the DQN agent with the CnnPolicy
model = PPO('CnnPolicy', env, verbose=1, tensorboard_log=f'tensorboard/{config["env_name"]}')

# model = PPO('CnnPolicy', env, verbose=0, buffer_size=100000, learning_starts=10000, target_update_interval=1000,
#             exploration_final_eps=0.05, tensorboard_log=f'{log_dir}/tensorboard/', train_freq=4, batch_size=32
#             , gradient_steps=1, exploration_fraction=.15)  # , exploration_fraction=.1 #, exploration_final_eps=0.02

# Train the model
model.learn(total_timesteps=1000000, callback=checkpoint_callback, progress_bar=True)

# Save the final model
model.save(f'{log_dir}/{config["env_name"]}_final')

# Test the trained model
# test_env = create_env(game_name, seed=1)
# obs = test_env.reset()
# done = False
# while not done:
#     action, _states = model.predict(obs, deterministic=True)
#     obs, reward, done, info = test_env.step(action)
#     test_env.render()

# test_env.close()
```
